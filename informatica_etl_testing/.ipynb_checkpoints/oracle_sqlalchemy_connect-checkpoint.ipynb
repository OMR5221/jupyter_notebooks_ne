```python
import xml.etree.ElementTree as ET
# create element tree object
xmlFile = 'ewis-battne_BMY_BATTxxx_SUB0001_52M2_Breaker_Position_10302018-10312018.xml'

ns = {'fpl_pgen': '{http://fpl.pgen/}',
      'fpl_pgen2': 'http://fpl.pgen',
      'soap_env': 'http://schemas.xmlsoap.org/soap/envelope'}


tree = ET.parse(xmlFile)
# get root element 
root = tree.getroot()
print('ROOT TAG:' + str(root.tag))
print('ROOT ATTRIB:' + str(root.attrib))
print('ROOT TEXT:' + str(root.text))
    
    
# tags = [elem.tag for elem in root.iter()]
# print(tags)

for elem in root.iter(ns['fpl_pgen']+'Value'):
    print(elem.text)
    
for elem in root.findall('.//{http://fpl.pgen/}Value'):
    print(elem.text)
    
for elem in root.findall('.//{0}Value'.format(ns['fpl_pgen'])):
    print(elem.text)
```

```python
for point in root.findall('.//{0}PIPoint'.format(ns['fpl_pgen'])):
    print(point.attrib)
    
    for val in point.findall('.//{0}Values'.format(ns['fpl_pgen'])):
        print('Timestamps: ' + val.attrib['Time'])
```

```python

for child in root:
    print(child.tag, child.attrib)
    if child.tag == "{http://fpl.pgen/}Values":
        for step_child in child:
            print(step_child.tag)
```

```python
# iterate over the entire tree
print("-" * 40)
print("Iterating using a tree iterator")
print("-" * 40)
iter_ = tree.getiterator()
for elem in iter_:
    print("TAG: ",elem.tag)
    print("\tATTRIB: ",elem.attrib)
    print("\tTEXT: ", elem.text)
 
```
