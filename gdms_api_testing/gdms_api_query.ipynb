```python
import requests

r = requests.get("https://gdms.windlogics.com/api/site/", verify=False, proxies={"https": "https://webproxyjb.fpl.com:8080"})
print(r.text)
```

```python
# Making multiple requests with Sessions:
import requests

s = requests.Session()
s.proxies = {
                "http": "http://webproxyjb.fpl.com:8080",
                "https": "https://webproxyjb.fpl.com:8080"
            }

r = s.get("https://gdms.windlogics.com/api/site/", verify=False)
print(r.text)
```
